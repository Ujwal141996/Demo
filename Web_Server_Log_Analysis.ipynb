```python
import gzip
import re
import datetime as dt
from collections import defaultdict, Counter

LOG_FILE = "calgary_access_log.gz"

def main():
    logs = []

    pattern = re.compile(
        r'(\S+) (\S+) (\S+) \[([^\]]+)\] "([^"]*)" (\d{3}) (\S+)'
    )

    # ------------------------------
    # Part 1: Load & Clean Data
    # ------------------------------
    with gzip.open(LOG_FILE, 'rt', encoding='utf-8', errors='ignore') as f:
        for line in f:
            match = pattern.match(line)
            if not match:
                continue

            host, rfc931, authuser, timestamp, request, status, bytes_ = match.groups()

            try:
                date_obj = dt.datetime.strptime(
                    timestamp.split()[0],
                    "%d/%b/%Y:%H:%M:%S"
                )
            except:
                continue

            filename = None
            if request:
                parts = request.split()
                if len(parts) >= 2:
                    filename = parts[1]

            if bytes_ == "-":
                bytes_ = None
            else:
                bytes_ = int(bytes_)

            logs.append({
                "host": host,
                "date": date_obj.date(),
                "hour": date_obj.hour,
                "filename": filename,
                "status": int(status),
                "bytes": bytes_
            })

    # ------------------------------
    # Q1: Total log records
    # ------------------------------
    print("Q1 Total Records:", len(logs))

    # ------------------------------
    # Q2: Unique hosts
    # ------------------------------
    unique_hosts = len(set(log["host"] for log in logs))
    print("Q2 Unique Hosts:", unique_hosts)

    # ------------------------------
    # Q3: Date-wise unique filenames
    # ------------------------------
    date_filename = defaultdict(set)
    for log in logs:
        if log["filename"]:
            date_str = log["date"].strftime("%d-%b-%Y")
            date_filename[date_str].add(log["filename"])

    datewise_unique = {k: len(v) for k, v in date_filename.items()}
    print("Q3 Date-wise Unique Filenames:", datewise_unique)

    # ------------------------------
    # Q4: Count of 404 responses
    # ------------------------------
    count_404 = sum(1 for log in logs if log["status"] == 404)
    print("Q4 404 Count:", count_404)

    # ------------------------------
    # Q5: Top 15 filenames with 404
    # ------------------------------
    counter_404_files = Counter()
    for log in logs:
        if log["status"] == 404 and log["filename"]:
            counter_404_files[log["filename"]] += 1

    print("Q5 Top 15 404 Files:", counter_404_files.most_common(15))

    # ------------------------------
    # Q6: Top 15 file extensions with 404
    # ------------------------------
    counter_404_ext = Counter()
    for log in logs:
        if log["status"] == 404 and log["filename"]:
            if "." in log["filename"]:
                ext = log["filename"].split(".")[-1]
                counter_404_ext[ext] += 1

    print("Q6 Top 15 404 Extensions:", counter_404_ext.most_common(15))

    # ------------------------------
    # Q7: Total bandwidth per day for July 1995
    # ------------------------------
    bandwidth = defaultdict(int)
    for log in logs:
        if log["date"].year == 1995 and log["date"].month == 7:
            if log["bytes"]:
                date_str = log["date"].strftime("%d-%b-%Y")
                bandwidth[date_str] += log["bytes"]

    print("Q7 Bandwidth July 1995:", dict(bandwidth))

    # ------------------------------
    # Q8: Hourly request distribution
    # ------------------------------
    hourly = Counter()
    for log in logs:
        hourly[log["hour"]] += 1

    print("Q8 Hourly Distribution:", dict(hourly))

    # ------------------------------
    # Q9: Top 10 most requested filenames
    # ------------------------------
    filename_counter = Counter()
    for log in logs:
        if log["filename"]:
            filename_counter[log["filename"]] += 1

    print("Q9 Top 10 Filenames:", filename_counter.most_common(10))

    # ------------------------------
    # Q10: HTTP response code distribution
    # ------------------------------
    status_counter = Counter()
    for log in logs:
        status_counter[log["status"]] += 1

    print("Q10 Status Distribution:", dict(status_counter))


if __name__ == "__main__":
    main()
```
